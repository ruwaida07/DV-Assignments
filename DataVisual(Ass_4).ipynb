```python
from matplotlib import pyplot as plt
import numpy as np

cars = ['MARUTI', 'HYUNDAI', 'TATA',
		'KIA', 'MAHINDRA', 'RENAULT']

data = [50, 17, 7, 5, 5, 3]

fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars)

plt.show()
```

```python
import matplotlib.pyplot as plt
   
States = ['MH','KA','UP','KL','DL','AP','GA','WB','UK','SK']
Confirmed_Cases = [115494 , 37929 , 1697 , 113595 , 798 , 30300 ,  1995 , 15690 ,  1319 , 2177]
 
plt.bar(States, Confirmed_Cases)
plt.title('Cases per state ')
plt.xlabel('States')
plt.ylabel('Number of Confirmed cases in India ')
plt.show()
```
